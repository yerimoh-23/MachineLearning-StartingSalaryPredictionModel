```python
#Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier
```

```python
# read files
college = pd.read_csv('college.csv')
```

```python
# variables in college dataset
college.keys()
```

```python
n_samples, n_features = college.shape

print('Number of samples:', n_samples)
print('Number of features:', n_features)
```

```python
college
```

```python
# mean early career pay for public and private institutions
# comparison of the average early career pay between the two types of institutions

# Calculate mean early career pay by institution type
mean_early_career_pay = college.groupby('type')['early_career_pay'].mean()

# Plotting
plt.figure(figsize=(8, 6))

# Create barplot
mean_early_career_pay.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Mean Early Career Pay by Institution Type')
plt.xlabel('Institution Type')
plt.ylabel('Mean Early Career Pay')
plt.xticks(rotation=0)
plt.grid(axis='y')

plt.show()
```

```python
# relationship between the proportion of women in institutions and early career pay
plt.figure(figsize=(8, 6))

# Create scatter plot
plt.scatter(college['women_proportion'], college['early_career_pay'], color='purple', alpha=0.5)
plt.title('Proportion of Women vs Early Career Pay')
plt.xlabel('Proportion of Women')
plt.ylabel('Early Career Pay')
plt.grid(True)

plt.show()
```

```python
# relationship between the proportion of internationals in institutions and early career pay
plt.figure(figsize=(8, 6))

# Create scatter plot
plt.scatter(college['Foreign_proportion'], college['early_career_pay'], color='olivedrab', alpha=0.5)
plt.title('Proportion of Internationals vs Early Career Pay')
plt.xlabel('Proportion of Internationals')
plt.ylabel('Early Career Pay')
plt.grid(True)

plt.show()
```

```python
# relationship between the total enrollment in institutions and early career pay
plt.figure(figsize=(8, 6))

# Create scatter plot
plt.scatter(college['total_enrollment'], college['early_career_pay'], color='orange', alpha=0.5)
plt.title('Total Enrollment vs Early Career Pay')
plt.xlabel('Total Enrollment')
plt.ylabel('Early Career Pay')
plt.grid(True)

plt.show()
```

```python
# relationship between the STEM percentage in institutions and early career pay
plt.figure(figsize=(8, 6))

# Create scatter plot
plt.scatter(college['stem_percent'], college['early_career_pay'], color='indianred', alpha=0.5)
plt.title('STEM Percentage vs Early Career Pay')
plt.xlabel('STEM Percentage')
plt.ylabel('Early Career Pay')
plt.grid(True)

plt.show()
```

```python
# relationship between the net cost in institutions and early career pay
plt.figure(figsize=(8, 6))

# Create scatter plot
plt.scatter(college['net_cost'], college['early_career_pay'], color='cornflowerblue', alpha=0.5)
plt.title('Net Cost vs Early Career Pay')
plt.xlabel('Net Cost')
plt.ylabel('Early Career Pay')
plt.grid(True)

plt.show()
```

```python
# relationship between the in/out state total in institutions and early career pay
plt.figure(figsize=(8, 6))

# Create scatter plot
plt.grid()
plt.scatter(college['in_state_total'], college['early_career_pay'], color='forestgreen', alpha=0.5, label='in state')
plt.scatter(college['out_of_state_total'], college['early_career_pay'], color='gold', alpha=0.5, label='out of state')

plt.title('In/Out of State Total vs Early Career Pay')
plt.xlabel('In/Out of State Total')
plt.ylabel('Early Career Pay')

plt.legend(loc='upper left')

plt.show()
```
