```python
tuition_income = pd.read_csv('/content/drive/MyDrive/COMSC335/CourseProject/tuition_income.csv')
tuition_cost = pd.read_csv('/content/drive/MyDrive/COMSC335/CourseProject/tuition_cost.csv')
```

```python
tuition_income = tuition_income.dropna()
tuition_cost = tuition_cost.dropna()
```

```python
tuition_income
```

```python
tuition_cost
```

```python
tuition_cost = tuition_cost[['name', 'type','degree_length', 'in_state_total', 'out_of_state_total']]
```

```python
tuition_income = tuition_income[['name','net_cost']].groupby(['name']).first().reset_index()
```

```python
tuition = tuition_income.merge(right = tuition_cost, on = 'name')
tuition
```

```python
tuition['net_cost'] = tuition['net_cost'].round(2)
tuition
```

```python
tuition.to_csv('/content/drive/My Drive/tuition.csv', index=False)
```
